### 2. Data Exploratory Analysis (DEA)
  - How to load, construct and navigate in N-dimensional images using a computerized tomography images of human torax.

```python
import matplotlib.pyplot as plt
```

```python
def show_image(image:iio.core.util.Array) -> None:
  '''
  Shows image from imageIO array using matplotlib.pyplot
  Args:
    image (iio.core.util.Array): imageIO array
  '''
  plt.figure()
  plt.imshow(X=image)
```

```python
show_image(im_220)
show_image(im_221)
show_image(im_222)
show_image(im_224)
show_image(im_225)
```

An important detail in DICOM images is their metadata. So, this metadata can include:
  - patient demographics data: name, age, gender, clinical information
  - acquisition information: image format, sampling rates, data type, modality (such as X-ray, CT or MRI)

We can see our metadata by accessing [image].meta in ImageIO.

```python
im_220.meta.keys()
```
